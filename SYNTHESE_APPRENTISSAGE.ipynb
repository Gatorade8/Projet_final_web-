# Synthèse de tout ce qu'on a vu sur le projet Finance Zen

Salut, voilà un résumé de tout ce qu'on a abordé avec toutes tes questions. Ça te fera une bonne note de révision si tu oublies des trucs. J'ai tout regroupé par thèmes !

## 1. La configuration et les fichiers de base
- **package.json** : c'est la carte d'identité du projet. C'est NPM qui lit ça pour savoir quoi installer. Il y a aussi les scripts comme `lint` (qui lance eslint pour vérifier les erreurs) et `preview` (pour tester le build final en local).
- **eslint.config.js** : c'est le livre de règles de ton code. C'est le correcteur orthographique qui souligne en orange quand tu fais n'importe quoi (boucle infinie, variable pas utilisée).
- **vite.config.ts** : ça configure Vite. Le `import react from '@vitejs/plugin-react'` donne à Vite le pouvoir de comprendre le code React et active le rechargement automatique quand tu sauvegardes.
- **main.tsx** : c'est l'interrupteur général. Il prend ton composant App et l'injecte (render) dans le HTML. Le StrictMode c'est une ceinture de sécurité pour détecter les bugs au développement.

## 2. React et le front-end
- **.tsx au lieu de .jsx** : le T c'est pour TypeScript. Ça permet d'avoir un filet de sécurité, l'ordinateur te prévient direct si tu oublies un champ ou si tu envoies du texte au lieu d'un nombre.
- **Composants** : `App.tsx` c'est le grand patron (le cerveau). Il contient les données et donne les ordres. Les autres (TransactionForm, etc.) sont les employés. Les 3 panneaux de solde sont direct dans App car c'est juste des calculs rapides faits par le patron.
- **Props** : c'est les données ou les fonctions qu'on passe d'un parent à un enfant dans les balises (genre `onSubmit={handleAddTransaction}`). Ça permet à l'enfant de remonter les informations au parent en retour (le tuyau d'évacuation).
- **Chart.js** : on doit mettre les codes couleurs JS dans App.tsx et pas dans le css car le graphique dessine dans un canvas HTML et il ne lit pas les variables css. On utilise `useRef` (la pince) pour garder le contrôle du canvas et on détruit toujours l'ancien graphique avant d'en dessiner un nouveau dans le `useEffect`.
- **Les Handlers (handleAddTransaction...)** : ils font toujours deux choses. Ils appellent le serveur API avec un fetch pour sauvegarder, et font un dispatch (`useReducer`) pour que React mette à jour l'écran en direct.

## 3. Le Back-end (`server.js` et données)
- **transactions.json** : c'est juste le frigo (fichier texte bête). Ce n'est pas lui qui force la forme des données, c'est TypeScript dans le front-end.
- **fileURLToPath / __dirname** : comme on est en type module, la variable globale `__dirname` n'existe plus. On doit la recréer pour dire au serveur où chercher le dossier data localement peu importe d'où on lance l'application.
- **Middleware (`app.use...`)** : c'est le péage sur l'autoroute de l'API. `express.json()` traduit le texte brut en objet javascript, `cors()` donne le droit à ton React de parler avec le serveur Node.
- **fs (File System)** : le module de Node.js pour interagir avec le disque dur (`readFileSync`, `writeFileSync`). Sans lui rien n'est sauvegardé une fois le pc éteint.
- **transactions vs newTransaction** : `transactions` au pluriel c'est le tableau complet (l'historique). `newTransaction` au singulier c'est juste le dernier élément reçu depuis le formulaire.
- **Routes GET, POST, DELETE** : ce sont les portes d'entrée de l'API. Le code React fait la requête, et c'est express.js qui l'écoute et interagit vraiment avec le fichier json après réception.

## 4. Le CSS et le design
- **[data-theme="dark"]** : les crochets servent à cibler un attribut HTML précis. Ça permet d'activer tout le mode sombre d'un seul coup plutôt que rajouter une classe partout.
- **\*** : sélecteur universel. Applique les règles à toute la page en même temps (souvent pour remettre les marges à zéro).
- **Les variables `--`** : définit des variables css. Très utile pour le thème, il suffit de changer la variable une fois pour que tout le site change.
- **outline: none** : supprime la bordure bleue moche par défaut des navigateurs quand on clique sur un input.
- **rem** : unité relative intelligente basée sur la taille de police du navigateur (1rem = souvent 16px). Ça améliore l'accessibilité pour le zoom.
- **Glassmorphisme** : l'alliance de `background: rgba(...)` (transparent) avec `backdrop-filter: blur(...)` (flou). Ça donne l'effet de vitre dépolie très classe qu'on voit sur le site.

## 5. Ta super grille de référence hyper détaillée

| Concept ou Terme | A quoi ça sert en gros | Pourquoi on fait ça ? (Le secret pro) |
| --- | --- | --- |
| **package.json** | La carte d'identité du projet, il liste les dépendances (comme React/Vite) et les scripts de commande. | Parce que NPM et Node.js ont besoin de ce fichier pour installer exactement les bons outils pour faire tourner ton code. |
| **ESLint** | Le flic de la syntaxe, il te crie dessus (en orange) si tes variables ne sont pas utilisées par exemple. | Pour garder un code ultra clean avant le test. Ça évite surtout les bugs débiles comme les faux noms de variables. |
| **.tsx (TypeScript)** | React mais en TypeScript. Grosse amélioration par rapport à du simple `.jsx`. | Ça ajoute un "filet de sécurité". L'éditeur t'oblige à dire le type exact (nombre, texte) pour que l'app ne plante jamais à cause d'un mauvais format. |
| **StrictMode** | Une alerte développeur activée dans le `main.tsx`. | Ça s'assure que tes fonctions n'ont pas d'impacts étranges, il déclenche exprès le code 2 fois pour voir s'il est strict. |
| **Props** | Info ou cadeau qu'un parent donne à son enfant, comme `<BudgetChart theme={theme} />`. | Obligatoire pour communiquer ! Sans ça, les fichiers seraient isolés. Ça rend ton design souple et réutilisable ! |
| **onSubmit={...}** | C'est une prop "fonction". Le parent prête une de ses méthodes à l'enfant. | C'est le tuyau d'évacuation : l'enfant appelle ça pour dire "tiens patron j'ai fini prends les données" car seul le parent sait parler à Node.js. |
| **useRef** | La pince qui permet à React de rattraper un tag HTML (souvent le `<canvas>`). | Parce que React veut normalement tout contrôler. Mais comme Chart.js dessine à la main dessus, il faut lui donner un accès prioritaire pour poser ses couleurs ! |
| **===** | Égalité stricte ! Vérifie la valeur ET le type (5 === 5) -> Vrai \| (5 === "5") -> Faux. | On l'utilise au lieu du `==` normal, ça sauve de pleins de problèmes car c'est très précis et ne te laisse pas faire un mélange de types accidentel ! |
| **? : (Ternaire)** | Une écriture raccourcie du if/else sur une seule ligne : `question ? vrai : faux`. | Extrêmement utile partout en React pour définir à l'écran de quelle façon l'interface doit s'afficher (mode sombre ou clair rapidement). |
| **Middleware** | `app.use` comme un péage sur l'autoroute. C'est l'étape entre React et ta route finale dans le back-end. | Par exemple, un middleware (`express.json()`) déchiffre tout ce qui a été envoyé en texte brut et le transforme en objet Javascript utilisable ! |
| **fs** | File System. Module de Node pour accéder véritablement aux dossiers et fichiers système. | Indispensable pour la base de données : ça permet d'opérer sur `transactions.json` pour qu'il garde en mémoire long-terme pour les dépenses de demain. |
| **__dirname** | Chemin local complet de ton projet sur ton pc (ex C:/user/...). | Comme on utilise les modules récents, cette variable n'existe plus automatiquement, donc il faut le coder à la main pour dire où est le dossier data. |
| **transaction vs newTransaction** | `transactions` est la liste globale (Le livre des comptes). `newTransaction` est le petit nouveau reçu (Ticket de caisse). | Très important car ton back-end va prendre ton ticket pour l'ajouter (`unshift()`) au début de la liste totale et sauvegarder le livre. |
| **GET, POST, DELETE** | Les méthodes principales dans l'API et les routes. | Un ordre de communication : React envoie POST pour ajouter, DELETE pour détruire. Le server.js sait alors quoi faire selon la route appelée ! |
| **Handlers (`App.tsx`)** | Les chefs d'orchestre de tes clics de boutons genre 'handleDelete'. | Ils font toujours 2 étapes : l'appel vers l'API (`fetch`) pour écrire sur le disque du pc, et un appel de liste (dispatch) pour que tout soit vu à l'écran immédiatement. |
| **\* (CSS) & [data-theme]** | L'étoile = toutes les balises pour reset les marges. L'attribut [data-theme] cible le mode littéralement. | L'étoile empêche des marges indésirables et l'attribut permet de passer un seul CSS global plutôt qu'éditer chaque bouton un par un lors du changement de thème. |
| **--variables** | Pour créer un nom de base comme `--bc-color:red`. | S'utilise avec `var()`. Super pratique parce que pour le mode sombre on édite que ce fichier avec 2 sets de variables sans toucher au reste des classes dans tes formulaires par exemple ! |
| **Glassmorphisme** | Effet visuel de vitre : `background: rgba(...)` mi-clair + `backdrop-filter`. | Ça rajoute de la classe, en floutant ce qu'il y a derrière tes panels pour donner l'impression qu'ils sont transparents sans tuer ta visibilité ! |
| **rem, et outline:** | rem l'unité relative intelligente; et null pour le contour outline. | Les px sont fixes, les rem s'adaptent si l'utilisateur zoome sur son navigateur. Et pour l'outline on regarde le focus propre pour ne pas avoir la ligne bleue par défaut. |

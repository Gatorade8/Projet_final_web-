## But de l'exercice

L'objectif de cet exercice est de creer une **application web complète de gestion de budget personnel**.
On apprend ici a faire communiquer un **frontend React** avec un **backend Express.js**.
C'est ce qu'on appelle une architecture **client-serveur**.

### Ce qu'on pratique ici :
- Créer une interface React avec TypeScript
- Gérer un état global avec `useReducer` (pattern Redux-like)
- Créer une API REST avec Express.js
- Persister les données dans un fichier JSON (comme une mini base de données)
- Réaliser des appels HTTP (`fetch`) entre le front et le back
- Afficher des graphiques avec Chart.js
- Implémenter un système de theme clair/sombre

---

## Architecture globale du projet

```
gestionnaire-budget/
│
├── src/                   (FRONTEND - code React)
│   ├── App.tsx            (composant principal, toute la logique)
│   ├── index.css          (tous les styles CSS)

└── index.html             (page HTML racine)
```

### Schema de communication

```
[Navigateur / React]  ---HTTP fetch--->  [Vite Proxy]  --->  [Express server.js]
     port 5173                                                      port 3001
                                                                         |
                                                              lit/ecrit dans
                                                           data/transactions.json
```

Le **proxy Vite** permet d'eviter les problemes de CORS en dev : quand le front fait
un appel vers `/api/...`, Vite redirige automatiquement vers `localhost:3001`.

# FICHIER 2 : `vite.config.ts`

**Role** : Configure l'outil de build Vite et met en place un **proxy** pour
relier le front et le back en developpement.

```typescript
import { defineConfig } from 'vite'
import react from '@vitejs/plugin-react'

export default defineConfig({
    plugins: [react()],      // active le support JSX/TSX
    server: {
        proxy: {
            '/api': 'http://localhost:3001'
        }
    }
})
```

### Pourquoi le proxy ?

Sans le proxy, si le front (sur le port 5173) tente d'appeler le back (sur le port 3001),
le navigateur bloque la requete a cause de la **politique CORS** (Cross-Origin Resource Sharing).

Avec le proxy, Vite intercepte tous les appels commencant par `/api` et les
redirige vers `http://localhost:3001`. Du point de vue du navigateur, tout
vient du meme serveur : pas de CORS.

```
fetch('/api/transactions')  ->  Vite intercepte  ->  http://localhost:3001/api/transactions
```

**Role** : C'est la **base de donnees** de l'application. Un simple fichier JSON que Express


### Structure d'une transaction

| Champ | Type | Description |
|---|---|---|
| `id` | string | Identifiant unique genere cote client |
| `name` | string | Description de la transaction |
| `amount` | number | Montant en euros (toujours positif) |
| `type` | `"revenu"` ou `"depense"` | Indique si c'est une entree ou une sortie d'argent |
| `category` | string | Categorie parmi 8 choix possibles |
| `date` | string | Date au format ISO `YYYY-MM-DD` |

### Avantages et limites de cette approche

**Avantages** :
- Simple a mettre en place, pas besoin d'installer une vraie base de donnees
- Le fichier est lisible directement dans n'importe quel editeur

**Limites** :
- Pas de requetes complexes (pas de filtre SQL, pas de jointures)
- Pas adapte a plusieurs utilisateurs en meme temps (risque de conflit d'ecriture)
- Pour un vrai projet on utiliserait SQLite, PostgreSQL ou MongoDB

## 4.2 — La reconstruction de `__dirname` en ESM

```javascript
const __filename = fileURLToPath(import.meta.url)
const __dirname = path.dirname(__filename)
```

En JavaScript avec la syntaxe **CommonJS** (`require`), la variable `__dirname`
existe automatiquement et contient le dossier du fichier courant.

Mais avec **ESM** (`import/export`), cette variable n'existe plus.
On doit la reconstruire manuellement :
- `import.meta.url` donne l'URL du fichier courant (ex: `file:///C:/projet/server.js`)
- `fileURLToPath()` la convertit en chemin normal (`C:\projet\server.js`)
- `path.dirname()` isole le dossier (`C:\projet`)

## 4.3 — Les fonctions utilitaires read/write

```javascript
function readTransactions() {
    try {
        const data = fs.readFileSync(DATA_FILE, 'utf-8')
        return JSON.parse(data)     // on convertit le texte JSON en tableau JS
    } catch (err) {
        return []                   // si le fichier n'existe pas -> tableau vide
    }
}

function writeTransactions(transactions) {
    fs.writeFileSync(
        DATA_FILE,
        JSON.stringify(transactions, null, 4),   // null = pas de remplacement
        'utf-8'                                  // 4 = indentation de 4 espaces
    )
}
```

Ces deux fonctions sont le coeur du backend. Au lieu d'une vraie base de donnees,
on lit et on ecrit directement dans le fichier JSON.
Le `try/catch` dans `readTransactions` empeche le serveur de planter si le fichier
n'existe pas encore.

## 4.4 — Les 4 routes de l'API

### GET /api/transactions
```javascript
app.get('/api/transactions', (req, res) => {
    const transactions = readTransactions()
    res.json(transactions)    // repond avec le tableau JSON
})
```
Simple : on lit le fichier et on le renvoie en JSON. Code HTTP 200 par defaut.

### POST /api/transactions
```javascript
app.post('/api/transactions', (req, res) => {
    const transactions = readTransactions()
    const newTransaction = req.body          // le corps de la requete (JSON)

    if (!newTransaction.id) {
        newTransaction.id = Date.now().toString() + Math.random().toString(36).substring(2)
    }
    if (!newTransaction.date) {
        newTransaction.date = new Date().toISOString().split('T')[0]
    }

    transactions.unshift(newTransaction)     // ajoute en DEBUT de tableau (plus recent en premier)
    writeTransactions(transactions)          // sauvegarde
    res.status(201).json(newTransaction)     // 201 = Created
})
```

### DELETE /api/transactions/:id
```javascript
app.delete('/api/transactions/:id', (req, res) => {
    const transactions = readTransactions()
    // on garde toutes les transactions SAUF celle avec l'id en parametre
    const filtered = transactions.filter(t => t.id !== req.params.id)

    if (filtered.length === transactions.length) {
        // longueur identique = aucune suppression = l'id n'existait pas
        return res.status(404).json({ error: 'Transaction pas trouvee' })
    }

    writeTransactions(filtered)
    res.json({ message: 'Transaction supprimee' })
})
```

### DELETE /api/transactions (reset total)
```javascript
app.delete('/api/transactions', (req, res) => {
    writeTransactions([])    // on ecrit un tableau vide -> tout efface
    res.json({ message: 'Toutes les transactions ont ete supprimees' })
})
```

## 6.2 — Le Reducer (lignes 40 a 83)

Le reducer est au coeur de la gestion d'etat. C'est un **pattern inspire de Redux** :
au lieu de modifier l'etat directement, on "dispatche" des actions.

```typescript
type BudgetAction =
    | { type: 'ADD_TRANSACTION';   payload: Transaction }   // ajouter
    | { type: 'DELETE_TRANSACTION'; payload: string }       // supprimer (par id)
    | { type: 'LOAD_TRANSACTIONS'; payload: Transaction[] } // charger depuis le back
    | { type: 'RESET_TRANSACTIONS' }                        // tout vider

function budgetReducer(state: BudgetState, action: BudgetAction): BudgetState {
    switch (action.type) {
        case 'ADD_TRANSACTION':
            // on met la nouvelle transaction en PREMIER (ordre chronologique inverse)
            return { transactions: [action.payload, ...state.transactions] }

        case 'DELETE_TRANSACTION':
            // on filtre pour garder tout SAUF la transaction avec cet id
            return {
                transactions: state.transactions.filter(t => t.id !== action.payload)
            }

        case 'LOAD_TRANSACTIONS':
            // on remplace completement le state avec les donnees du serveur
            return { transactions: action.payload }

        case 'RESET_TRANSACTIONS':
            return { transactions: [] }
    }
}
```

### Pourquoi useReducer plutot que useState ?

| `useState` | `useReducer` |
|---|---|
| Simple, pour 1-2 valeurs | Ideal quand l'etat est complexe |
| La logique est dans le composant | La logique est centralisee dans le reducer |
| Difficile a tester | Facile a tester (pure function) |

Ici, on a une liste de transactions avec 4 operations differentes : `useReducer`
est le bon choix pour garder le code lisible.

```typescript
useEffect(() => {
```

La fonction `loadTransactions` (lignes 51-60) :
```typescript
async function loadTransactions(): Promise<Transaction[]> {
    try {
        const response = await fetch('/api/transactions')
        const data: Transaction[] = await response.json()
        return data
    } catch {
        return []     // si le serveur est eteint -> tableau vide
    }
}
```

Flux complet au demarrage :
1. React monte le composant `App`
2. Le `useEffect` se declenche
3. `fetch('/api/transactions')` envoie une requete GET
4. Vite redirige vers `http://localhost:3001/api/transactions`
5. Express lit `data/transactions.json` et repond
6. Le frontend dispatch `LOAD_TRANSACTIONS` avec les donnees
7. React re-render avec les transactions affichees

## 6.5 — Calculs du solde (lignes 137 a 153)

```typescript
// calcule le total des revenus avec reduce
const totalRevenu = state.transactions
    .filter(t => t.type === 'revenu')           // 1. garde les revenus
    .reduce((sum, t) => sum + t.amount, 0)      // 2. additionne tous les montants

const totalDepense = state.transactions
    .filter(t => t.type === 'depense')
    .reduce((sum, t) => sum + t.amount, 0)

const solde = totalRevenu - totalDepense
```

Le `reduce` fonctionne comme un accumulateur :
- `sum` commence a `0`
- Pour chaque transaction, on ajoute son montant a `sum`
- A la fin, `sum` contient le total

````python
# simulation du calcul de solde en Python

transactions = [
    {"name": "Salaire",    "amount": 2200, "type": "revenu"},
    {"name": "Loyer",      "amount": 650,  "type": "depense"},
    {"name": "Courses",    "amount": 85.5, "type": "depense"},
    {"name": "Freelance",  "amount": 350,  "type": "revenu"},
    {"name": "Metro",      "amount": 75.2, "type": "depense"},
    {"name": "Cinema",     "amount": 12,   "type": "depense"},
]

revenus  = [t["amount"] for t in transactions if t["type"] == "revenu"]
depenses = [t["amount"] for t in transactions if t["type"] == "depense"]

total_revenu  = sum(revenus)
total_depense = sum(depenses)
solde = total_revenu - total_depense

print(f"Total revenus  : +{total_revenu:.2f} EUR")
print(f"Total depenses : -{total_depense:.2f} EUR")
print(f"Solde          :  {solde:.2f} EUR")
````

## 6.7 — Les handlers CRUD (lignes 247 a 307)

### handleSubmit — Ajouter une transaction
```typescript
async function handleSubmit(e: React.FormEvent) {
    e.preventDefault()    // empeche le rechargement de la page (comportement par defaut des forms)

    if (!name.trim() || !amount || parseFloat(amount) <= 0) return   // validation

    const newTransaction: Transaction = {
        id: Date.now().toString() + Math.random().toString(36).substring(2),  // id unique
        name: name.trim(),
        amount: parseFloat(parseFloat(amount).toFixed(2)),  // arrondi a 2 decimales
        type,
        category,
        date: new Date().toISOString().split('T')[0],  // YYYY-MM-DD
    }

    // 1. envoie au serveur
    await fetch(API_URL, {
        method: 'POST',
        headers: { 'Content-Type': 'application/json' },
        body: JSON.stringify(newTransaction),
    })

    // 2. met a jour le state local (pas besoin de recharger depuis le serveur)
    dispatch({ type: 'ADD_TRANSACTION', payload: newTransaction })

    // 3. vide le formulaire
    setName('')
    setAmount('')
}
```

### Pourquoi on met a jour le state local ET le serveur ?

On pourrait simplement appeler `loadTransactions()` apres chaque modification
pour recharger tout depuis le serveur. Mais c'est plus lent.

Ici on adopte le pattern **optimistic update** :
- On met directement a jour l'interface (rapide, pas d'attente)
- On envoie aussi la requete au serveur (pour persister)
- Si le serveur echoue, l'interface reste quand meme coherente

## 6.8 — Le systeme de theme (lignes 99 a 128)

```typescript
const [theme, setTheme] = useState<'light' | 'dark'>(() => {
    const saved = localStorage.getItem('budget_theme')
    return (saved === 'dark') ? 'dark' : 'light'
})
```

L'initialisation du state utilise une **fonction** au lieu d'une valeur directe.
C'est ce qu'on appelle un "lazy initializer" : la fonction n'est executee
qu'une seule fois au montage, pas a chaque render.

```typescript
useEffect(() => {
    document.documentElement.setAttribute('data-theme', theme)
    localStorage.setItem('budget_theme', theme
}, [theme])
```

Quand le theme change :
1. On ajoute `data-theme="dark"` ou `data-theme="light"` sur la balise `<html>`
2. En CSS, la selector `[data-theme="dark"]` change toutes les variables CSS
3. On sauvegarde le choix dans `localStorage` pour le retrouver au prochain chargement

En CSS (`index.css`) :
```css
:root {
    --bg-primary: #f5efe6;    /* theme clair par defaut */
}
[data-theme="dark"] {
    --bg-primary: #1a1410;    /* override pour le theme sombre */
}
```

```css
@media (max-width: 700px) {
    .main-grid {
        grid-template-columns: 1fr;   /* 1 colonne sur mobile */
    }
}
```

Sur mobile (ecran < 700px), le formulaire et le graphique passent en colonne
verticale au lieu d'etre cote a cote.

# Comment lancer le projet

```bash
# dans le dossier gestionnaire-budget/

# installer les dependances (a faire une seule fois)
npm install

# Terminal 1 : lancer le backend Express
npm run server
# -> Serveur demarre sur http://localhost:3001

# Terminal 2 : lancer le frontend Vite
npm run dev
# -> App disponible sur http://localhost:5173
```

> IMPORTANT : il faut que les deux terminaux soient actifs en meme temps.
> Si le backend n'est pas lance, l'app frontend fonctionnera quand meme
> mais les transactions ne seront pas sauvegardees (tableau vide a chaque rechargement).

# Tester l'API directement

On peut tester l'API backend sans passer par l'interface React :

```bash
# recuperer toutes les transactions
curl http://localhost:3001/api/transactions

# ajouter une transaction
curl -X POST http://localhost:3001/api/transactions \
     -H "Content-Type: application/json" \
     -d '{"name":"Test","amount":42,"type":"depense","category":"loisirs"}'

# supprimer une transaction par son id
curl -X DELETE http://localhost:3001/api/transactions/MON_ID

# tout effacer
curl -X DELETE http://localhost:3001/api/transactions
```
